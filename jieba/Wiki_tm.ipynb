```python
import requests
from bs4 import BeautifulSoup as bs
import re
from operator import itemgetter
```

```python
def removeWord(text):
    rem = re.sub('[\\\\[\]\{\},.:;!?()\-+*=><#。、（），：」「^《》！『』/；【 】·“”？－…→%\n\t年的月日與有是於]','',text)
    return rem
```

```python
url = 'https://zh.wikipedia.org/wiki/%E8%94%A1%E8%8B%B1%E6%96%87'
res = requests.get(url)
soup = bs(res.text)
con = soup.select('#bodyContent')[0].text
rem = removeWord(con)     
```

```python
import jieba
jieba.set_dictionary('dict.txt.big')
jieba.add_word('蔡英文')
jieba.add_word('民進黨')
jieba.add_word('小英')
jieba.add_word('蔡主席')
jieba.add_word('臺灣')
jieba.add_word('劉憶如')
dic = {}
for ele in jieba.cut(rem):
    if not ele in dic:
        dic[ele] = 1
    else:
        dic[ele] = dic[ele] + 1
```

```python
#姓名的詞性代號為 nr,這邊多加一個無詞性 x 因為jieba裡面沒有蔡英文。指定詞性找出分詞中屬於這兩種詞性的詞
import jieba.analyse
tags = jieba.analyse.extract_tags(rem, topK=100, withWeight=False, allowPOS=('nr','x'))
print(tags)
```

```python
#將序列與已經做好分詞的字典做比對，只挑出詞性為nr,x的詞,並列出出現頻率
for ele in tags:
    if ele in dic:
        print(ele +'\t' +str(dic[ele]))
```

```python
def date_conversion(text):
    rem = re.sub('[\t]','',text)
    #將所有符合 0000年00月00日 的字串找出，用 0000-00-00替代
    year_day = re.compile(r'(\d+)年(\d+)月(\d+)日')
    d1 =  year_day.sub(r'\1-\2-\3',rem)
    #將所有符合 0000年00月 的字串找出，用 0000-00替代
    year_month = re.compile(r'(\d+)年(\d+)月')
    d2 = year_month.sub( r'\1-\2',d1)
   
    #頁面上日期格式都會變成 0000-00-00
    
    print(d2)
```

```python
date_conversion(con)
```
