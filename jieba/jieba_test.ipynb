```python
import jieba
jieba.set_dictionary('dict.txt.big')
seg =  "頂新味全二十一項油品混入「大統長基」的「銅葉綠素」黑心油案，台北地院審理近一年半在今天下午宣判。前頂新、味全董事長魏應充\
，依「詐欺罪」判刑4年。前頂新總經理「常梅峰」判刑3年10個月，前味全總經理「張教華」被判無罪。「味全公司」則被科罰金1千550萬元。\
全案仍可上訴。檢方指出，味全公司自89年起委由頂新代工，生產小包裝的家庭食用油，看準高價橄欖油、葡萄籽油等調合油市場，為降低原料成本、\
提升獲利，96年起夥同頂新向「大統長基」購買添加「銅葉綠素」等低價混充油。雖然「頂新」驗出大統的油品都有「脂肪酸組成異常」及「油脂色澤不自然」\
的情形，但在當時的董事長魏應充、總經理常梅峰低價採購油品的放水指示下，仍繼續向大統長基購買混充油，再由味全代工製成健康廚房橄欖油、\
葡萄籽油等油品供貨到各大通路販售。檢方依涉犯詐欺、偽造文書、食品衛生管理法等罪嫌，起訴魏應充、張教華等13人以及味全公司。"
```

### 全模式

```python
seg_list = jieba.cut(seg, cut_all=True)
print("Full Mode: " + "/ ".join(seg_list))
```

### 精確模式

```python
seg_list = jieba.cut(seg, cut_all=False)
print("Default Mode: " + "/ ".join(seg_list))
```

### 搜索引擎模式

```python
seg_list = jieba.cut_for_search(seg)  # 搜索引擎模式
print(", ".join(seg_list))
```

### 列出字頻

```python
import re
def removeWord(text):
    rem = re.sub('[\\\\[\]\{\},.:;!?()\-+*=><#。、（），：」「^《》！『』/；【 】·“”？－…→%\n\t年的月日與有是於]','',text)
    return rem
```

```python
rem = removeWord(seg)
dic ={}
for ele in jieba.cut(rem):
    if not ele in dic:
        dic[ele] = 1
    else:
        dic[ele] = dic[ele] + 1
```

```python
from operator import itemgetter
swd = sorted(dic.items(), key = itemgetter(1), reverse=True)
for ele in swd:
    print(ele[0],ele[1])
```

### 關鍵詞抽取

```python
import jieba.analyse
#此處只抽取名詞及人名
tags = jieba.analyse.extract_tags(rem, topK=100, withWeight=False, allowPOS=('nr','n'))
print(tags)
```
