```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#===================
#分類標籤
labels = np.array(['面向1','面向2','面向3',
                  '面向4','面向5','面向6'])
#共6類
dataLenth = 6
#數據
data = np.array([1,4,3,6,4,8])
#===================

angles = np.linspace(0, 2*np.pi, dataLenth, endpoint=False)
data = np.concatenate((data,[data[0]]))
angles = np.concatenate((angles,[angles[0]]))

fig = plt.figure()
ax = fig.add_subplot(111,polar=True)
ax.plot(angles, data, 'bo-',linewidth=2)
ax.fill(angles,data, facecolor='#0000ff',alpha=0.25)
ax.set_thetagrids(angles * 180/np.pi, labels, fontproperties="SimHei")
ax.set_title('求才分析',va='bottom',fontproperties="SimHei")
ax.set_rlim(0,10)
ax.grid(True)
plt.show()
```

```python
def drawphase(data,name,color):
    import numpy as np
    import matplotlib.pyplot as plt
    %matplotlib inline
    labels = np.array(['面向1','面向2','面向3','面向4','面向5','面向6'])
    dataLenth = 6
    
    
    angles = np.linspace(0, 2*np.pi, dataLenth, endpoint=False)
    data = np.concatenate((data,[data[0]]))
    angles = np.concatenate((angles,[angles[0]]))
    
    fig = plt.figure()
    ax = fig.add_subplot(111,polar=True)
    ax.plot(angles, data, 'bo-',linewidth=2)
    ax.fill(angles, data, facecolor=color,alpha=0.25)
    ax.set_thetagrids(angles * 180/np.pi, labels, fontproperties="SimHei")
    ax.set_title(name,va='bottom',fontproperties="SimHei")
    ax.set_rlim(0,10)
    ax.grid(True)
    plt.show()
```

```python
#test
a=3
b=5
c=4
d=2
e=7
f=1
data = [a,b,c,d,e,f]
drawphase(data,"TEST","#ff0055")
```
